```python
kws = [
    "решение дифференциальных уравнений частных производных",
    "системы дифференциальных уравнений",
    "системы дифференциальных уравнений общая теория"
]
```

```python
dsts = []

for kw1 in kws:
    for kw2 in kws:
        if kw1 != kw2:
            dsts.append(cosine(
                kw_to_emb(kw1),
                kw_to_emb(kw2)
            ))
```

```python
import numpy as np

np.mean(dsts)
```
